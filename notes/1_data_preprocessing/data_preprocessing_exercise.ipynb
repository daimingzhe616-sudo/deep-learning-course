创建包含更多行和列的原始数据集。
1. 删除缺失值最多的列。 
2. 将预处理后的数据集转换为张量格式

```python
import pandas as pd
import torch

data = pd.DataFrame({
    "A": [1, 2, None, 4, None],
    "B": [None, None, None, 3, 4],
    "C": [5, 6, 7, 8, 9],
    "D": [None, 2, None, 4, 5]
})

print(data)
```

```python
print(data.isna().sum()) #每列缺失值数量
```

```python
max_missing_col = data.isna().sum().idxmax() #缺失值最多的列
data = data.drop(columns=[max_missing_col])

print(data)
```

为了转换为张量，处理NaN，用均值填充

```python
data = data.fillna(data.mean())
```

```python
tensor_data = torch.tensor(data.values)

print(tensor_data)
```

如果我们将权重初始化为零，会发生什么。算法仍然有效吗？

如果权重都相同，那么在训练过程中所有神经元的梯度都会相同，每一层的神经元学习的参数就会相同

假设试图为电压和电流的关系建立一个模型。自动微分可以用来学习模型的参数吗? 

```python
import torch

I = torch.tensor([1.,2.,3.,4.])
V = torch.tensor([2.,4.,6.,8.])  # 假设R=2

w = torch.randn(1, requires_grad=True)

for _ in range(100):
    pred = w * I
    loss = ((pred - V)**2).mean()

    loss.backward()

    with torch.no_grad():
        w -= 0.01 * w.grad
        w.grad.zero_()

print(w)
```

从而优化温度。

为什么在squared_loss函数中需要使用reshape函数？ 

```python
def squared_loss(y_hat, y):
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

y_hat和y的形状经常不一样，因为每批量的每个样本输出一个预测值，模型输出y_hat一般是列向量，但数据集中取出来的标签y是标量，二者形状不一样

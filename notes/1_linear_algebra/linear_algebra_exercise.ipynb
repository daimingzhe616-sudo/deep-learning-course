```python
import torch
X = torch.arange(24).reshape(2, 3, 4)
len(X)
```

len(X)返回的是张量最外层的长度，也就是第0轴的大小。对于任意张量X，len(X)总是等于第0轴的长度

```python
A = torch.arange(20).reshape(5, 4)
A / A.sum(axis=1)
```

A的形状是（5，4），求和的形状是（5，），不能进行广播，直接相除会报错，改正：

```python
A / A.sum(axis=1, keepdims=True)
```

```python
A.shape
```

```python
A.sum(axis=1, keepdims=True).shape
```

此时的形状可以进行广播为（5，4），能进行逐元素相除

```python
X.sum(axis=0)
```

```python
X.sum(axis=1)
```

```python
X.sum(axis=2)
```

没有特别指定dim，那么torch.linalg.norm(X)会把整个张量看成一个“大向量”，然后计算它的范数。
对于多维张量，这等价于把所有元素平方后求和，再开根号。，最后输出一个标量

```python
X = torch.arange(24, dtype=torch.float32).reshape(2,3,4)
torch.linalg.norm(X)
```
